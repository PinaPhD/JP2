```python
#Creating the variables from Column "VALUE II" tagged x
ax = 35/y2s   #controller failure rate
bx = 13/y2s  #switch failure rate
cx = 2500/h2s    #controller failure detection rate
dx = 2500/h2s     #switch migration rate
ex = 2/h2s        #controller repair rate
fx = 360000/h2s        #switch failure detection rate
gx = 2/h2s        #switch repair rate
hx = 0.95      #probability of successful controller failure detection
ix = 0.95     #probability of successful switch migration
jx = 0.95      #probability of successful switch failure detection
kx = 0.95     #probability of successful PCEP trigger
mx = 0.95     #probability of redetection of switch failure
nx = 0.95     #probability of redetection of controller failure
px = 3.5      #repair rate scale from failed state to full operation
```

```python
#Constructing the transition rate matrix
Q_valueI = np.array([[-(h+j),0,0,0,h,0,j,0,0],
            [e,-(e+h),0,0,h,0,0,0,0],
            [g,0,-(k+g),0,0,0,0,0,0],
            [(p*e),0,0,-(p*e),0,0,0,0,0],
            [0,0,0,0,-(f+d),c,0,0,d],
            [n,c,0,0,0,-(c),0,0,0],
            [0,0,b,0,0,0,-(b+f),f,0],
            [m,0,0,0,0,0,0,-(m),0],
            [0,i,0,a,0,0,0,0,-(a)]])

# Create the DataFrame from Q_valueI
QI_df = pd.DataFrame(Q_valueI, columns=['W1', 'W2', 'W3', 'F', 'detC' , 'fdetC' , 'detS' , 'fdetS' , 'migS'],
                    index=['W1', 'W2', 'W3', 'F', 'detC' , 'fdetC' , 'detS' , 'fdetS' , 'migS'])

print("Transition Rate Matrix or Jump Rate Matrix")
print(QI_df) 
```

```python
#Constructing the transition rate matrix
Q_valueII = np.array([[-(hx+jx),0,0,0,h,0,j,0,0],
            [e,-(ex+hx),0,0,hx,0,0,0,0],
            [gx,0,-(kx+gx),0,0,0,0,0,0],
            [(p*ex),0,0,-(p*ex),0,0,0,0,0],
            [0,0,0,0,-(fx+dx),cx,0,0,d],
            [nx,cx,0,0,0,-(cx),0,0,0],
            [0,0,bx,0,0,0,-(bx+fx),fx,0],
            [mx,0,0,0,0,0,0,-(mx),0],
            [0,ix,0,ax,0,0,0,0,-(ax)]])

# Create the DataFrame from Q_valueI
QII_df = pd.DataFrame(Q_valueII, columns=['W1', 'W2', 'W3', 'F', 'detC' , 'fdetC' , 'detS' , 'fdetS' , 'migS'],
                    index=['W1', 'W2', 'W3', 'F', 'detC' , 'fdetC' , 'detS' , 'fdetS' , 'migS'])

print("Transition Rate Matrix or Jump Rate Matrix")
print(QII_df) 
```

```python
def calculate_steady_state_probabilities(Q):
    """
    Calculate steady-state probabilities for a Markov model.
    Args:
        Q (numpy.ndarray): Transition rate matrix.
    Returns:
        numpy.ndarray: Steady-state probabilities.
    """
    n = Q_valueI.shape[0]
    A = np.vstack((Q_valueI.T[:-1, :], np.ones(n)))
    b = np.array([0] * (n - 1) + [1])
    pi = solve(A, b)
    return pi / np.sum(pi)

def calculate_mtbf_mttr_mttf(Q_valueI, failure_state, repair_state):
    """
    Calculate MTBF, MTTR, and MTTF for a Markov model.
    Args:
        Q (numpy.ndarray): Transition rate matrix.
        failure_state (int): Index of the failure state.
        repair_state (int): Index of the repair state.
    Returns:
        tuple: MTBF, MTTR, MTTF
    """
    steady_state_probabilities = calculate_steady_state_probabilities(Q_valueI)
    
    failure_index = failure_state - 1  # Convert to 0-based index
    repair_index = repair_state - 1    # Convert to 0-based index
    
    mtbf = 1 / np.sum(steady_state_probabilities[:-1] * Q_valueI[:-1, failure_index])
    mttr = 1 / Q_valueI[failure_index, repair_index]
    mttf = 1 / np.sum(steady_state_probabilities[:-1] * Q_valueI[:-1, failure_index])

    return mtbf, mttr, mttf


# Example failure state and repair state
failure_state = 2
repair_state = 3

# Calculate steady-state probabilities, MTBF, MTTR, and MTTF
steady_state_probabilities = calculate_steady_state_probabilities(Q_valueI)
mtbf, mttr, mttf = calculate_mtbf_mttr_mttf(Q_valueI, failure_state, repair_state)

# Print results
print("Steady-State Probabilities:", steady_state_probabilities)
print("MTBF:", mtbf)
print("MTTR:", mttr)
print("MTTF:", mttf)
```
